# One Hot Encoding using with Sklearn

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
```

```python
cat_col= ['Model','Colour','Rear Camera','Front Camera','Processor_']
dummy= pd.get_dummies(n_df,columns=cat_col,dtype=int)
```

```python
dummy
```

# Split the dataset into training and testing sets (80% train, 20% test)

```python
x_train,x_test,y_train,y_test= train_test_split(dummy.drop('Price',axis=1),dummy['Price'],test_size=0.2,random_state=42)
```

```python
#n_df['Price']=n_df['Price'].str.replace(',','').astype(float)
selected_columns = ['Memory','RAM','Battery_' ,'Price']

# Creating a correlation matrix for selected columns
corre = n_df[selected_columns].corr()

#Visualize relationships between selected features and target variable
plt.figure(figsize=(8, 6))
sns.heatmap(corre, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap for Selected Columns')
plt.show()
```

```python
# Evaluate the model's performance
mse = mean_squared_error(y_test, y_pred)
```
